```
from os import listdir
from torchvision.transforms.functional import resize
from PIL import Image

outPath = "/content/drive/MyDrive/CS17 project/data/train_512HR/"

def ResizeAndSaveLRImages(dataset_dir):
  for x in listdir(dataset_dir):
    print("Resizing %s ..." % x)
    currentImage = Image.open(dataset_dir + x)
    resizedImage = resize(currentImage, 512, interpolation=Image.NEAREST)
    image_filename = x
    resizedImage.save(outPath + image_filename)


ResizeAndSaveLRImages("/content/drive/MyDrive/CS17 project/data/train_HR/")
```

```
from os import listdir
from torchvision.transforms.functional import resize
from PIL import Image
import time

# https://pillow.readthedocs.io/en/latest/handbook/concepts.html#PIL.Image.NEAREST

dir = "/content/drive/MyDrive/CS17 project/data/test/00003.png"
testImage = Image.open(dir)
outdir = "/content/drive/MyDrive/CS17 project/data/test"

# NEAREST
start = time.clock()
SRImage = resize(testImage, 224, interpolation=Image.NEAREST)
duration = time.clock() - start
print("NEAREST duration is:", str(duration))
SRImage.save(outdir + 'NEAREST_78_224' + '.png' )

# BOX
start = time.clock()
SRImage = resize(testImage, 224, interpolation=Image.BOX)
duration = time.clock() - start
print("BOX duration is:", str(duration))
SRImage.save(outdir + 'BOX_78_224' + '.png' )

# BILINEAR
start = time.clock()
SRImage = resize(testImage, 224, interpolation=Image.BILINEAR)
duration = time.clock() - start
print("BILINEAR duration is:", str(duration))
SRImage.save(outdir + 'BILINEAR_78_224' + '.png' )

# HAMMING
start = time.clock()
SRImage = resize(testImage, 224, interpolation=Image.HAMMING)
duration = time.clock() - start
print("HAMMING duration is:", str(duration))
SRImage.save(outdir + 'HAMMING_78_224' + '.png' )

# BICUBIC
start = time.clock()
SRImage = resize(testImage, 224, interpolation=Image.BICUBIC)
duration = time.clock() - start
print("BICUBIC duration is:", str(duration))
SRImage.save(outdir + 'BICUBIC_78_224' + '.png' )

# LANCZOS
start = time.clock()
SRImage = resize(testImage, 224, interpolation=Image.LANCZOS)
duration = time.clock() - start
print("LANCZOS duration is:", str(duration))
SRImage.save(outdir + 'LANCZOS_78_224' + '.png' )
```

```
from os import listdir
from torchvision.transforms.functional import resize
from PIL import Image

outPath = "/content/drive/My Drive/data/train_BlurHR/"

def ResizeAndSaveBlurHRImages(dataset_dir):
  for x in listdir(dataset_dir):
    print("Blurring %s ..." % x)
    HRImage = Image.open(dataset_dir + x)
    LRImage = resize(HRImage, 256, interpolation=Image.NEAREST)
    BlurHRImage = resize(LRImage, 2048, interpolation=Image.NEAREST)
    image_filename = x
    BlurHRImage.save(outPath + image_filename)

ResizeAndSaveBlurHRImages("/content/drive/My Drive/data/train_HR/")
```
